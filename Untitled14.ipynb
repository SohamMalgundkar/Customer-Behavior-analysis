```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.rename(columns={"purchase_amount_(usd)" : "purchase_amount"})
```

```python
df.columns
```

```python
# Create Age Group Column
df['age_group'] = pd.qcut(df['age'], q=4, labels=['Young', 'Adult','Middle-Aged', 'Senior'])
```

```python
df[['age','age_group']].head(5)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
'Fortnightly': 14,
'Weekly': 7,
'Monthly': 30,
'Quarterly': 90,
'Bi-Weekly': 14,
'Annually': 365,
'Every 3 Months': 90
}
df['purchase_frequency_days' ] = df['frequency_of_purchases'].map(frequency_mapping)

```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head()
```
